```python
# Importar librerías necesarias
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler

# Cargar los datasets limpios
movies_df = pd.read_csv('D:/Henry/Proyectos/Proyecto_Individual_1/data/movies_dataset_transformed.csv')
credits_df = pd.read_csv('D:/Henry/Proyectos/Proyecto_Individual_1/data/filtered_credits.csv')

# Mostrar las primeras filas para verificar los datos
print("Primeras filas de 'movies_dataset_transformed.csv':")
print(movies_df.head())

print("\nPrimeras filas de 'filtered_credits.csv':")
print(credits_df.head())

# Preprocesamiento: Normalizar la columna 'popularity'
scaler = StandardScaler()
movies_df['popularity_scaled'] = scaler.fit_transform(movies_df[['popularity']])

# Crear la matriz de características
features = movies_df[['popularity_scaled']]

# Calcular la matriz de similitud usando cosine similarity
similarity_matrix = cosine_similarity(features)

# Crear un DataFrame de similitud para fácil acceso
similarity_df = pd.DataFrame(similarity_matrix, index=movies_df['title'], columns=movies_df['title'])

# Función para recomendar películas
def recomendar_peliculas(titulo_pelicula, top_n=5):
    if titulo_pelicula not in similarity_df.index:
        return "Película no encontrada"
    
    # Obtener la similitud de la película especificada con todas las demás
    similar_scores = similarity_df[titulo_pelicula]
    
    # Ordenar las películas por similitud y seleccionar las más similares
    similar_scores = similar_scores.sort_values(ascending=False)
    top_similar_movies = similar_scores.index[1:top_n+1]  # Excluye la propia película
    
    return top_similar_movies.tolist()

# Probar la función de recomendación con un ejemplo
titulo_pelicula_ejemplo = 'The Shawshank Redemption'  # Cambia este título por el de tu elección
recomendaciones = recomendar_peliculas(titulo_pelicula_ejemplo)
print(f"\nPelículas recomendadas similares a '{titulo_pelicula_ejemplo}':")
print(recomendaciones)


```
